```
from statistics import stdev,mean,variance
import numpy as np
import sklearn

class ListObject():
  def __init__(self,length):
    self.length=length
    self.value=None
    self.array=None
  
  def build_array(self, value):
    self.value=value
    self.array=[value for i in range(self.length)]
    return self.array

  def updating(self,last_value):
    self.value=last_value

    self.array.pop(0)
    self.array.append(last_value)

    return self.array


class ListsGenerator:
  def __init__(self,arrays_lengths):
    self.arrays_lengths=arrays_lengths
    self.last_value=None
    self.array_ListObject=None

  def generator(self,value):
    self.last_value=value
    self.array_ListObject=[]
    
    for i in self.arrays_lengths:
      self.array_ListObject.append(ListObject(i))
      self.array_ListObject[-1].build_array(value)
    
    return self.array_ListObject

  def data_income(self,last_value):
    self.last_value=last_value

    for i in self.array_ListObject:
      i.updating(last_value)

    return self.array_ListObject

class DataGenerator:
  def __init__(self,array_lengths):
    """
    Per array length , two values will be returned (variance and mean)
    Parameters
    ----------
    array_length: python list
      This list wil be used for determine the lengths of the lists for the processing

    """
    self.value=None
    self.array_lengths=array_lengths
    self.GeneratorObj = ListsGenerator(array_lengths)
    self.array_ListObject=None
    self.results=None

  def data_input(self,last_value):
    """
    Here the data comes and generates the correspondent array values
    """
    if self.value == None:
      self.array_ListObject=self.GeneratorObj.generator(last_value)

    else:
      self.array_ListObject=self.GeneratorObj.data_income(last_value)

    self.value=last_value

    return self._preparing_values()

  def _preparing_values(self,):
    """
    Values are calculated from the different arrays
    """
    self.results={}
    aux_length=len(self.array_lengths)

    for i in range(aux_length):
      self.results['variance_len'+str(self.array_lengths[i])]=variance(self.array_ListObject[i].array)
      self.results['mean_len'+str(self.array_lengths[i])]=mean(self.array_ListObject[i].array)

    return self.results
```
