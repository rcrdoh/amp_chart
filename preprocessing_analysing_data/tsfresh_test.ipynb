# Preparacion de datos (4)

```
## preparcion de nombres de columnas para concatenar mas adelante
names=[]
for i in range(4):
  names.append('time_'+str(i))
  names.append('current_'+str(i))
```

```
names
```

```
data=pd.read_csv(path+'sintetic_data4.csv',names=names)
```

```
data.tail()
```

```
series_data=pd.DataFrame()
for i in range(4):
  temp_dataframe=data[['time_'+str(i),'current_'+str(i)]]
  temp_dataframe=temp_dataframe.rename(columns={'time_'+str(i):'time','current_'+str(i):'current'})
  temp_dataframe['id'] = pd.Series([i+1 for j in range(len(temp_dataframe))])
  series_data=pd.concat([series_data,temp_dataframe],ignore_index=True)
```

```
series_data
```

```
extracted_features = tsfresh.extract_features(series_data, column_id='id',column_sort="time")
```

```
extracted_features
```

```
series_data100=pd.DataFrame()
for i in range(0,12,2):
  temp_dataframe=data100[[i,i+1]]
  temp_dataframe=temp_dataframe.rename(columns={i:'time',i+1:'current'})
  #temp_dataframe['id'] = pd.Series([i+1 for j in range(len(temp_dataframe))])
  series_data100=pd.concat([series_data100,temp_dataframe],ignore_index=True)
```

```
series_data100
```

```
def id_per_serie(dataseries,length_subseries):
  """
  creates a id column , where gives an id per subseries in the dataseries

  OBS: the subseries inside dataseries have the same length

  Returns
  -------
  numpy array:
    array(1D) with IDs corresponding to every different subseries

  """
  extra_column=np.array([])
  
  for i in range(int(len(dataseries)/length_subseries)):
    temp_series=[int(i+1) for j in range(length_subseries)]
    extra_column=np.append(extra_column,temp_series)
  
  extra_column=extra_column.astype('int32')

  return extra_column
```
